# Loan sanction Prediction using SVM Classifier

In this notebook, following concepts are explored

#Missing values are filled using fillna method
#get_dummies method is used to categorize the categorical variables
#sklearn labelencoding is used to map different labels
#sklearn standard scaler is used to perform feature normalization
#Data is trained using SVM classifier
#Predictions are generated fro the new test data using the trained classifier model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train=pd.read_csv('train.csv')
```

```python
train
```

```python
y=train['Loan_Status']
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
Y=le.fit_transform(y)
Y
```

```python
train.isnull().sum()
```

```python
train.fillna(method='bfill',inplace=True)
```

```python
train.isnull().sum()
```

```python
d1=pd.get_dummies(train[['Gender','Married','Dependents',
                      'Education','Self_Employed',
                      'Property_Area']],drop_first=True)
```

```python
train['S_term']=train['Loan_Amount_Term'].map(lambda a: 1 if a >=12 and a <120 else 0)
train['M_term']=train['Loan_Amount_Term'].map(lambda b: 1 if b >=120 and b <300 else 0)
train['L_term']=train['Loan_Amount_Term'].map(lambda c: 1 if c >=300 else 0)
```

```python
d3=pd.get_dummies(train['Credit_History'],drop_first=True)
```

```python
d0=train.drop(['Loan_ID','Gender','Married','Dependents',
                      'Education','Self_Employed',
                      'Property_Area','Loan_Amount_Term','Credit_History','Loan_Status'],axis=1)
d0
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
d0=sc.fit_transform(d0)
d0=pd.DataFrame(d0)
d0
```

```python
df=pd.concat([d0,d1,d3],axis=1)
df
```

```python
Y
```

# Training accuracy

```python
from sklearn.svm import SVC
```

```python
model=SVC(C=1,gamma=1,kernel='linear')
```

```python
model.fit(df,Y)
```

```python
Y_pred=model.predict(df)
```

```python
from sklearn.metrics import accuracy_score
```

```python
score=accuracy_score(Y_pred,Y)*100
score
```
